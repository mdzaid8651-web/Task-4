
# Data Cleaning Project using a Kaggle Dataset

This notebook demonstrates a complete **data cleaning process** using the famous **Titanic Dataset** from Kaggle.

Dataset Source: Titanic Dataset from Kaggle  
The notebook covers:
- Loading dataset
- Checking missing values
- Handling null values
- Removing duplicates
- Fixing data types
- Handling outliers
- Renaming columns
- Saving cleaned data


## Step 1: Import Libraries


```

import pandas as pd
import numpy as np
```


## Step 2: Load Dataset

You can download the Titanic dataset from Kaggle:
https://www.kaggle.com/c/titanic/data

After downloading, place `train.csv` in the same folder as this notebook.


```

# Load dataset

df = pd.read_csv('train.csv')

# Display first 5 rows
df.head()
```


## Step 3: Basic Information


```

df.info()
```

```

df.shape
```

```

df.describe()
```


## Step 4: Check Missing Values


```

df.isnull().sum()
```


## Step 5: Handle Missing Values


```

# Fill missing Age values with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill missing Embarked values with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin column because it has many missing values
df.drop(columns=['Cabin'], inplace=True)

# Remove rows with missing Fare values
df.dropna(subset=['Fare'], inplace=True)

df.isnull().sum()
```


## Step 6: Check Duplicate Records


```

print("Duplicate Rows:", df.duplicated().sum())
```

```

# Remove duplicates
df.drop_duplicates(inplace=True)

print("Duplicate Rows After Cleaning:", df.duplicated().sum())
```


## Step 7: Fix Data Types


```

df.dtypes
```

```

# Convert Survived column to category type
df['Survived'] = df['Survived'].astype('category')

df.dtypes
```


## Step 8: Rename Columns


```

df.rename(columns={
    'Pclass': 'Passenger_Class',
    'Sex': 'Gender'
}, inplace=True)

df.head()
```


## Step 9: Handle Outliers


```

# Detect outliers using IQR method

Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
```

```

# Remove outliers

df = df[(df['Fare'] >= lower_limit) & (df['Fare'] <= upper_limit)]

df.shape
```


## Step 10: Final Cleaned Dataset


```

df.head()
```


## Step 11: Save Cleaned Dataset


```

df.to_csv('cleaned_titanic_dataset.csv', index=False)

print("Cleaned dataset saved successfully.")
```
